```python
import pandas as pd
import numpy as np

np.random.seed(42)

dates = pd.date_range(start="2023-01-01", end="2023-12-31", freq="D")
n = 500

data = {
    "date": np.random.choice(dates, n),
    "city": np.random.choice(["Lagos", "Abuja", "Kano"], n),
    "category": np.random.choice(["Electronics", "Clothing", "Food", "Pharmacy"], n),
    "quantity": np.random.randint(1, 20, n),
    "unit_price": np.random.uniform(5, 500, n).round(2),
    "customer_age": np.random.randint(18, 70, n),
    "payment_method": np.random.choice(["Cash", "Card", "Transfer"], n),
}

# Introduce missing values
df = pd.DataFrame(data)
df.loc[np.random.choice(df.index, 30), "customer_age"] = np.nan
df.loc[np.random.choice(df.index, 20), "payment_method"] = np.nan

print(df.shape)
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure 'month' is formatted as month name for better readability
df['month_name'] = df['date'].dt.strftime('%b') # Abbreviated month name

# Calculate total sales by month and category
monthly_category_sales = df.groupby(['month_name', 'category'])['total_sales'].sum().unstack()

# Define the order of months for plotting
month_order = pd.CategoricalDtype([
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
], ordered=True)
monthly_category_sales = monthly_category_sales.reindex(month_order.categories)

plt.figure(figsize=(15, 7))
monthly_category_sales.plot(kind='bar', ax=plt.gca())
plt.title('Total Sales by Category and Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
