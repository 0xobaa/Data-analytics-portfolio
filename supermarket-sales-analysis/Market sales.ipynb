```python
import pandas as pd
import numpy as np

date_range_2= pd.date_range(start = "2022-01-01", periods = 24, freq = "ME")
reported_cases = [63,56,34,56,87,90,34,56,78, 55,48,90,46,35,87,65,56,47,12,19,20,43,40,15]

my_df = pd.DataFrame({
    "date": date_range_2,
    "reported_cases": reported_cases
})
print(my_df)
```

```python
 my_df["month"] = my_df["date"].dt.month
 my_df["year"] = my_df["date"].dt.year
 my_df["quarter"] = my_df["date"].dt.quarter
 print(my_df['month'])
 print(my_df['year'])
 print(my_df['quarter'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
plt.fill_between(my_df["date"], my_df["reported_cases"], alpha=0.1, color="steelblue")
sns.lineplot(data=my_df, x="date", y="reported_cases", marker="o")
plt.axhline(y=my_df["reported_cases"].mean(), color="red", linestyle="--")
plt.title ("Yearly Reported Cases")
plt.ylabel ("Reported Cases")
plt.xlabel ("Year")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure 'month' is formatted as month name for better readability
df['month_name'] = df['date'].dt.strftime('%b') # Abbreviated month name

# Calculate total sales by month and category
monthly_category_sales = df.groupby(['month_name', 'category'])['total_sales'].sum().unstack()

# Define the order of months for plotting
month_order = pd.CategoricalDtype([
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
], ordered=True)
monthly_category_sales = monthly_category_sales.reindex(month_order.categories)

plt.figure(figsize=(15, 7))
monthly_category_sales.plot(kind='bar', ax=plt.gca())
plt.title('Total Sales by Category and Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
